## Testes para validar o funcionamento dos algoritmos de SSSP

Bloco de testes para os algoritmos de Single Source Shortests Path

```python
# Grafo de exemplo utilizado para todos os testes
example_graph = {
    'A': [('B', 1.5), ('C', 2.0), ('E', 2.8)],
    'B': [('D', 3.0), ('F', 1.7)],
    'C': [('D', 1.2), ('F', 2.5), ('G', 3.1)],
    'D': [('H', 2.2)],
    'E': [('F', 1.1), ('I', 2.9)],
    'F': [('H', 1.4), ('J', 2.6)],
    'G': [('J', 1.3)],
    'H': [('K', 2.0)],
    'I': [('J', 1.8)],
    'J': [('K', 1.5), ('L', 2.7)],
    'K': [('M', 1.9)],
    'L': [('M', 2.3)],
    'M': []
}

# n - Número de vértices
n = len(example_graph)
# m - Número de arestas
m = sum(len(neighbors) for neighbors in example_graph.values())
```

#### 1. Testando os algoritmos convencionais

Os dois métodos convencionais (Dijkstra e Bellman-Ford) são utilizados para compor a resolução proposta por Duan et al. No caso, Bellman-Ford é utilizado para realizar o relaxamento dos vértices da fronteira S, com o intuito de diminuí-la, e uma versão limitada do Dijkstra é utilizada para finalizar os vértices determinados pelo bmssp

Vamos tomar a solução do Dijkstra como verdade para ser base de comparação

```python
# Adicionando o caminho para o módulo de grafos
import sys
sys.path.append('C:/Users/Pichau/bcc/tcc_graphs')

from graphs import dijkstra

resultado_dijkstra = dijkstra(example_graph, 'A')
```

Testando para o algoritmo de Bellman-Ford

```python
from graphs import bellman_ford

resultado_bellman_ford = bellman_ford(example_graph, 'A')

assert resultado_dijkstra ==  resultado_bellman_ford, "Os resultados do Dijkstra e Bellman-Ford devem ser iguais para grafos sem arestas de peso negativo."
```

#### 2. Aplicando para o BMSSP

```python
from duan_et_al import get_params, bmssp
from math import ceil, log2

# Calculando os parâmetros para o algoritmo de Duan et al.
k, t = get_params(n)

# Definindo a origem
s = 'A'

# Inicializando as estimativas de distância mínima
d_hat = {v: float('inf') for v in example_graph}
d_hat[s] = 0

# Parâmetros para a chamada de topo
log_n = log2(n) if n > 1 else 1
l_inicial = ceil(log_n / t)
B_inicial = float('inf')
S_inicial = {s}

bmssp(l_inicial, B_inicial, S_inicial, example_graph, d_hat, k, t)
print(d_hat)
print(resultado_dijkstra)
```
